# 2. PDF and CDF

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
```

```python
x = np.linspace(-6, 6, 100)
```

```python
fx = norm.pdf(x, loc=0, scale=1) # the location is mean, the scale is standard deviation
```

```python
plt.plot(x, fx);
```

```python
Fx = norm.cdf(x, loc=0, scale=1)
```

```python
plt.plot(x, Fx);
```

```python
logfx = norm.logpdf(x, loc=0, scale=1)
```

```python
plt.plot(x, logfx);
```
