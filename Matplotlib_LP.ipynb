# 3. Scatterplot

```python
X = np.random.randn(100, 2)
plt.scatter(X[:,0], X[:,1]);
```

```python
X = np.random.randn(200, 2)
X[0:50] += 3 # To simulate classification
Y = np.zeros(200) # To assign color: default color
Y[:50] = 1 # To assign color: will be Y for first 50 points
plt.scatter(X[:,0], X[:,1], c=Y);
```

# 6. Exercise

```python

X = np.random.uniform(low=-1, high=1, size=(1000, 2))

def dist_xor(element):
  x0 = 0 if element[0] < 0 else 1
  x1 = 0 if element[1] < 0 else 1
  return x0 ^ x1

Y = [dist_xor(element) for element in X]
plt.scatter(X[:,0], X[:,1], c=Y);

```
