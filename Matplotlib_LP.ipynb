# 2. Line Chart

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 1000)
```

```python
y = np.sin(x) + 0.2 * x
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.show() # required if without Notbook
```

```python
plt.plot(x, y)
plt.xlabel("input")
plt.ylabel("output")
plt.title("my plot"); # ; remove return string
```

# 3. Scatterplot

```python
X = np.random.randn(100, 2)
plt.scatter(X[:,0], X[:,1]);
```

```python
X = np.random.randn(200, 2)
X[0:50] += 3 # To simulate classification
Y = np.zeros(200) # To assign color: default color
Y[:50] = 1 # To assign color: will be Y for first 50 points
plt.scatter(X[:,0], X[:,1], c=Y);
```
